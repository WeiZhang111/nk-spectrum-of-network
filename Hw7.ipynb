### Question 1
Create the n-k connectivity spectrum (exactly for low k and sampling for high k) for all power girds and discuss.

Please notice that for reducing computational demanding, the sampled combination number is selected as 20. Noise appears in the spectrums

Power grid 1:

```python
mat1 = scipy.io.loadmat('1.mat')
A1 = mat1['A']
s1 = np.shape(A1)
G1 = nx.DiGraph() 
for i in range(s1[0]): 
    for j in range(s1[1]): 
        if A1[i,j] == 1: 
            G1.add_edge(i,j)        
G1 = nx.DiGraph.to_undirected(G1)
e=[(i+1) for i in range(G1.number_of_nodes())]
```

```python
# loop for solving spectrum

Spec1 = []
for n in e:
    IsConnected = []

    RemovalList = []                             
    for ii in range(50):                                   # 50 combinations are sampled, for making it computationally cheaper
        random_list=random.sample(e,n)
        RemovalList.append(random_list)
            
    k = 0
    while k < len(RemovalList):
        mat1 = scipy.io.loadmat('1.mat')
        A1 = mat1['A']
        s1 = np.shape(A1)
        G1 = nx.DiGraph() 
        for i in range(s1[0]): 
            for j in range(s1[1]): 
                if A1[i,j] == 1: 
                    G1.add_edge(i,j)        
        G1 = nx.DiGraph.to_undirected(G1)

        G1.remove_nodes_from(RemovalList[k])
        if nx.is_connected(G1):

            IsConnected.append('a')

        k += 1
    
    
    Spec1.append(len(IsConnected)/len(RemovalList))
    del(RemovalList)
```

```python
plt.plot(e,Spec1)
plt.title('Spectrum for power grid 1')
plt.xlabel('# of removal nodes')
plt.ylabel('Probability of fully connected')
plt.show()
```
